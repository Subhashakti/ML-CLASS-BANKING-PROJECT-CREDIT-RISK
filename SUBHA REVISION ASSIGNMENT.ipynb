```python
#Observations- - 1. SeriousDlqin2yrs- Discrete variable- Min Value- 0 Max value-24
 
#Output-  All result into 0 & 1. Data is without any issue & can be used for analysis.
```

```python
print(loans['age'].value_counts())
sns.countplot(y="age",  data=loans)
```

```python
#Observation- 2. age variable- Discrete in this case.- Min- > 18 year( Major) eligibility for loan & Max- age -90

#Output- here we find:
 1. #age- 0- Not possible
 2. #age- > 100- unlikely 
    
```

```python
#Step-2- Data Cleaning- Since this is discrete, Numerical, we can find average/Media & replace the age > 90 with medain

loans['age'].median()
```

```python
#Imputations with Median - For Age Variable

#Imputation with median
loans['age_new']= loans['age']
loans['age_new'] [loans['age'] > 90]= loans['age'].median()
print(loans['age_new'].value_counts())
loans['age_new'] [loans['age'] < 1]= loans['age'].median()
print(loans['age_new'].value_counts())
```

```python
#Observation on age_new criteria- Now age is clean. No '0' age & > 90 years age person.
```

```python
print (loans ['NumberOfTime30-59DaysPastDueNotWorse'].value_counts())
sns.countplot(y="NumberOfTime30-59DaysPastDueNotWorse", data=loans)
```

```python
#Observations- In all the default bracket, there is 98 & 96 with same no of borrower.
#3. NumberOfTime30-59DaysPastDueNotWorse- Discrete variable- min 0 & max 24.
#Imputation with median

loans['DPD0']= loans['NumberOfTime30-59DaysPastDueNotWorse']
loans['DPD0'][loans['NumberOfTime30-59DaysPastDueNotWorse'] > 24] = loans['NumberOfTime30-59DaysPastDueNotWorse'].median
print(loans['DPD0'].value_counts())

#Output- NOW 1st DPD 
#This is clean & between 0 to <=24.
```
